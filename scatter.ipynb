# 2.2 插入一列，名称为color
1,根据species的值，设定color列的值如下

Iris-setosa  --> 0

Iris-versicolor -->1

Iris-virginica -->2

2，查看首部5行

3，查看50-55行

4，查看尾部5行


```python
#write code here
def judge(x):
    if 'Iris-setosa' == x:
        return 0
    elif 'Iris-versicolor'== x:
        return 1
    elif 'Iris-virginica' == x:
        return 2
dataframe['color'] = dataframe.species.apply(lambda x: judge(x))
```

# 把所有的样本点用scatter显示
1,横坐标，从[sepal length,sepal width,petal length,petal width]选择一个属性

2,纵坐标，从[sepal length,sepal width,petal length,petal width]选择另一个属性

提示，也就是横坐标和纵坐标是[sepal length,sepal width,petal length,petal width]的两两组合，比如sepal length-sepal length，但是不要跟自己组合，比如sepal length-sepal length

3,散点颜色为1、2、3，分别表示Iris-setosa、Iris-versicolor、Iris-virginica

4,看看各个类别时间是否有比较清晰的分类边界，考察两个属性是否可以把各个类别分开

```python
#write code here
f, [[ax1, ax2, ax3],[ax4, ax5, ax6]] = plt.subplots(2, 3, figsize=(15, 8))
sns.set_style("whitegrid")

sns.scatterplot(dataframe['sepal length'],dataframe['sepal width'],hue=1+dataframe['color'],ax=ax1)

sns.scatterplot(dataframe['sepal length'],dataframe['petal length'],hue=1+dataframe['color'],ax=ax2)

sns.scatterplot(dataframe['sepal length'],dataframe['petal width'],hue=1+dataframe['color'],ax=ax3)

sns.scatterplot(dataframe['sepal width'],dataframe['petal length'],hue=1+dataframe['color'],ax=ax4)

sns.scatterplot(dataframe['sepal width'],dataframe['petal width'],hue=1+dataframe['color'],ax=ax5)

sns.scatterplot(dataframe['petal length'],dataframe['petal width'],hue=1+dataframe['color'],ax=ax6)

plt.tight_layout()


```
